```
import matplotlib.pyplot as plt
from matplotlib import lines
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.tree import  DecisionTreeClassifier
from sklearn.neighbors import  KNeighborsClassifier
from sklearn import  metrics
from numpy.core.umath import sign
import sys

# calculating Euclidean distance 
# using linalg.norm() from numpy
def EuclideanDistance(p, q):
    distance = np.linalg.norm(p-q)
    return distance

# custom KNN algorithm
def CustomKNN(DataX, DataY, k, testX):
    distance = []
    testY = []
    # calculate each data point distance and sort them
    for n in range(len(DataX)):
        distance.append([EuclideanDistance(DataX[n], testX), n])
    distance.sort()

    # find out K nearnest neighbor for test data and return the output value appear most times as result
    for i in range(k):
        testY.append(DataY[distance[i][1]])
    return max(set(testY), key = testY.count)

# load data and split them in X, y data sets
def loadData():
    digitData = load_digits(return_X_y=True)
    Xs = digitData[:-1]
    X = []
    for xs in Xs:
        for x in xs:
            X.append(x)
    y = digitData[-1]
    return X,y


# main function
if __name__ == '__main__':
    np.set_printoptions(threshold=sys.maxsize)
    # load digit data:
    X, y = loadData()
    
    # separate the data by using sklearn function
    X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, random_state=42, test_size=0.33)

    classifiers = [
        (DummyClassifier(strategy="most_frequent"), "Simple_Majority"),
        (DecisionTreeClassifier(criterion="entropy"), "Decision Tree"),
        (KNeighborsClassifier(n_neighbors=5), "5NN"),
        ((KNeighborsClassifier(n_neighbors=7), "7NN"))]

    # the colors of the plot (parameter c)
    # should represent the predicted class value
    # we found this linewidth to work well
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1, projection='3d')
   
    # you need to define the min and max values from the data
    x_min = 1
    x_max = 3 # run 3Trials
    y_min = 0
    y_max = 100 # max output value
    accuracy_max = 100 # max accuracy
    accuracy_min = 0
    step_size = 1
    xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, step_size),
                            np.arange(y_min, y_max, step_size),
                            np.arange(accuracy_min, accuracy_max, step_size))

    SimpleMajorityOutput = []
    DecisionTreeOutput = []
    KNN_5_Output = []
    KNN_7_Output = []
    CusKNN_1_Output = []
    CusKNN_3_Output = []
    CusKNN_5_Output = []
    CusKNN_7_Output = []
    CusKNN_9_Output = []

    SimpleMajorityAccuracy = 0.0
    DecisionTreeAccuracy = 0.0
    KNN_5_Accuracy = 0.0
    KNN_7_Accuracy =0.0
    CusKNN_1_Accuracy = 0.0
    CusKNN_3_Accuracy = 0.0
    CusKNN_5_Accuracy = 0.0
    CusKNN_7_Accuracy = 0.0
    CusKNN_9_Accuracy = 0.0

    SimpleMajorityAccuracyList = []
    DecisionTreeAccuracyList = []
    KNN_5_AccuracyList = []
    KNN_7_AccuracyList = []
    CusKNN_1_AccuracyList = []
    CusKNN_3_AccuracyList = []
    CusKNN_5_AccuracyList = []
    CusKNN_7_AccuracyList = []
    CusKNN_9_AccuracyList = []

    # run sklearn different machine learning algorithms for 3 times for average accuracy and output
    for i in range(3):
        CusKNNLables = []
        X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, random_state=42, test_size=0.33)
        for clf, name in classifiers:
            clf.fit(X_Train, y_Train)
            newlabels = clf.predict(X_Test)
            accuracy = metrics.accuracy_score(y_Test, newlabels)
            if name == "Simple_Majority":
                # if the algorithm is simple majority
                print("Simple classifier:")
                print("Accuracy: " + str(accuracy))
                print(newlabels)
                SimpleMajorityAccuracy+=accuracy*100
                SimpleMajorityOutput.append(newlabels)
                SimpleMajorityAccuracyList.append(accuracy)
            elif name == "Decision Tree":
                # if the algorithm is the decision tree
                print("Decision tree:")
                print("Accuracy: " + str(accuracy))
                print(newlabels)
                DecisionTreeAccuracy += accuracy*100
                DecisionTreeOutput.append(newlabels)
                DecisionTreeAccuracyList.append(accuracy)
            elif name == "5NN":
                # if the algorithm is KNN with K = 5
                print("5NN:")
                print("Accuracy: " + str(accuracy))
                print(newlabels)
                KNN_5_Accuracy += accuracy*100
                KNN_5_Output.append(newlabels)
                KNN_5_AccuracyList.append(accuracy)
            elif name == "7NN":
                # if the algorithm is KNN with K = 7
                print("7NN:")
                print("Accuracy: " + str(accuracy))
                print(newlabels)
                KNN_7_Accuracy += accuracy*100
                KNN_7_Output.append(newlabels)
                KNN_7_AccuracyList.append(accuracy)
        

        # run custom KNN with K = 1
        curCus1NNAccuracy = 0
        for curTest in X_Test:
            CusKNN_1_Output.append(CustomKNN(X_Train, y_Train, 1, curTest))

        # calculate custom 1NN accuracy
        for i in range(len(y_Test)):
            if CusKNN_1_Output[i] == y_Test[i]:
                curCus1NNAccuracy += 1
        CusKNN_1_Accuracy += curCus1NNAccuracy

        # print out the output for custom 1NN
        print("Custom 1NN:")
        print("Accuracy: "+ str(curCus1NNAccuracy/len(y_Test)))
        print(np.array(CusKNN_1_Output))
        CusKNN_1_AccuracyList.append(curCus1NNAccuracy/len(y_Test))        

        # run custom KNN with K = 3
        curCus3NNAccuracy = 0
        for curTest in X_Test:
            CusKNN_3_Output.append(CustomKNN(X_Train, y_Train, 3, curTest))

        # calculate custom 3NN accuracy
        for i in range(len(y_Test)):
            if CusKNN_3_Output[i] == y_Test[i]:
                curCus3NNAccuracy += 1
        CusKNN_3_Accuracy += curCus3NNAccuracy

        # print out the output for custom 3NN
        print("Custom 3NN:")
        print("Accuracy: " + str(curCus3NNAccuracy/len(y_Test)))
        print(np.array(CusKNN_3_Output))
        CusKNN_3_AccuracyList.append(curCus3NNAccuracy/len(y_Test))

        
        # run custom KNN with K = 5
        curCus5NNAccuracy = 0
        for curTest in X_Test:
            CusKNN_5_Output.append(CustomKNN(X_Train, y_Train, 5, curTest))

        # calculate custom 5NN accuracy
        for i in range(len(y_Test)):
            if CusKNN_5_Output[i] == y_Test[i]:
                curCus5NNAccuracy += 1
        CusKNN_5_Accuracy += curCus5NNAccuracy

        # print out the output for custom 5NN
        print("Custom 5NN:")
        print("Accuracy" + str(curCus5NNAccuracy/len(y_Test)))
        print(np.array(CusKNN_5_Output))
        CusKNN_5_AccuracyList.append(curCus5NNAccuracy/len(y_Test))

        # run custom KNN with K = 7
        curCus7NNAccuracy = 0
        for curTest in X_Test:
            CusKNN_7_Output.append(CustomKNN(X_Train, y_Train, 7, curTest))

        # calculate custom 7NN accuracy
        for i in range(len(y_Test)):
            if CusKNN_7_Output[i] == y_Test[i]:
                curCus7NNAccuracy += 1
        CusKNN_7_Accuracy += curCus7NNAccuracy

        # print out the output for custom 7NN
        print("Custom 7NN:")
        print("Accuracy: " + str(curCus7NNAccuracy/len(y_Test)))
        print(np.array(CusKNN_7_Output))
        CusKNN_7_AccuracyList.append(curCus7NNAccuracy/len(y_Test))

        # run custom KNN with K = 9
        for curTest in X_Test:
            CusKNN_9_Output.append(CustomKNN(X_Train, y_Train, 9, curTest))

        # calculate custom 9NN accuracy
        curCus9NNAccuracy = 0
        for i in range(len(y_Test)):
            if CusKNN_9_Output[i] == y_Test[i]:
                curCus9NNAccuracy+=1
        CusKNN_9_Accuracy += curCus9NNAccuracy

        # print out the output for custom 9NN
        print("Custom 9NN:")
        print("Accuracy: " + str(curCus9NNAccuracy/len(y_Test)))
        print(np.array(CusKNN_9_Output))
        CusKNN_9_AccuracyList.append(curCus9NNAccuracy/len(y_Test))


    # print out average accuracy for different algorithms
    print("Majority classifier average accuracy: " +str(SimpleMajorityAccuracy/3))  
    print("Decision Tree average accuracy: "+str(DecisionTreeAccuracy/3))
    print("5NN average accuracy: "+str(KNN_5_Accuracy/3))
    print("7NN average accuracy:" + str(KNN_7_Accuracy/3))
    print("Custom 1NN average accuracy:" + str(CusKNN_1_Accuracy/3/len(y_Test)*100))
    print("Custom 3NN average accuracy:" + str(CusKNN_3_Accuracy/3/len(y_Test)*100))
    print("Custom 5NN average accuracy:" + str(CusKNN_5_Accuracy/3/len(y_Test)*100))
    print("Custom 7NN average accuracy:" + str(CusKNN_7_Accuracy/3/len(y_Test)*100))
    print("Custom 9NN average accuracy:" + str(CusKNN_9_Accuracy/3/len(y_Test)*100))

    # draw Majority classifier line
    a = ax.scatter([1,2,3], SimpleMajorityAccuracyList, SimpleMajorityAccuracyList, c="black", marker='s', edgecolors='k', linewidth=0.2)
    # draw decision tree line
    b = ax.scatter([1,2,3], DecisionTreeAccuracyList, DecisionTreeAccuracyList, c="red", marker='d', edgecolors='k', linewidth=0.2)
    # System 5NN
    c = ax.scatter([1,2,3],KNN_5_AccuracyList, KNN_5_AccuracyList, c="blue", marker='^', edgecolors='k', linewidth=0.2)
    # System 7NN
    d = ax.scatter([1,2,3],KNN_7_AccuracyList, KNN_7_AccuracyList, c="blue", marker='8', edgecolors='k', linewidth=0.2)
    # custom 3NN
    e = ax.scatter([1,2,3], CusKNN_3_AccuracyList, CusKNN_3_AccuracyList, c="yellow", marker='o', edgecolors='k', linewidth=0.2)
    # custom 5NN
    f = ax.scatter([1,2,3], CusKNN_5_AccuracyList, CusKNN_5_AccuracyList, c="green", marker='D', edgecolors='k', linewidth=0.2)
    # custom 7NN
    h = ax.scatter([1,2,3], CusKNN_7_AccuracyList, CusKNN_7_AccuracyList, c="magenta", marker='p', edgecolors='k', linewidth=0.2)
    # custom 9NN
    g = ax.scatter([1,2,3], CusKNN_9_AccuracyList, CusKNN_9_AccuracyList, c="cyan", marker='*', edgecolors='k', linewidth=0.2)




    # you will want to enhance the plot with a legend and axes titles
    ax.set_xlabel("Trial Number")
    ax.set_ylabel("Accuracy")
    ax.set_zlabel("Accuracy")
    ax.legend((a,b,c,d,e,f,h,g),("Majority classifier", "decision tree", "5NN", "7NN", "Custom3NN", "Custom5NN", "Custom7NN", "Custom9NN"), scatterpoints = 1, loc="upper left")
    plt.show()
```
